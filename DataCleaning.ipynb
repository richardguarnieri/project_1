```python
#We have to change from demografic_df
# 'Czechia','China, Macao Special Administrative Region','Saint Vincent and the Grenadines',
#    'United Kingdom of Great Britain and Northern Ireland','United States of America'
# TO
#'Czech Republic', 'Macau', 'Saint Vincent and Grenadines',
#       'United Kingdom', 'United States'   
#We change the name to an easier one, or one that causes less confussion
demografic_df['Country'] = demografic_df['Country'].replace(['China, Macao Special Administrative Region'],'Macau')
demografic_df['Country'] = demografic_df['Country'].replace(['Saint Vincent and the Grenadines'],'Saint Vincent and Grenadines')
demografic_df['Country'] = demografic_df['Country'].replace(['United Kingdom of Great Britain and Northern Ireland'],'United Kingdom')

suicide_df['country'] = suicide_df['country'].replace(['Czech Republic'],'Czechia')
suicide_df['country'] = suicide_df['country'].replace(['United States'],'United States of America')
```
